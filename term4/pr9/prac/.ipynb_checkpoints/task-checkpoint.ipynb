```python
len_filt = starting_table[starting_table['Length'] < 500]
```

```python
len_filt = len_filt[len_filt['Length'] > 470]
```

```python
index_with_nan = len_filt.index[len_filt.isnull().any(axis=1)]
full_info = len_filt.copy()
full_info.drop(index_with_nan, 0, inplace=True)
```

```python
full_info
```

```python
len(np.unique(full_info['Taxonomic lineage (FAMILY)']))
```

```python
full_info_subset = full_info.copy()
full_info_subset[['Taxonomic lineage (GENUS)']].drop_duplicates(subset=['Taxonomic lineage (GENUS)'])
```

```python
full_info_subset
```

```python
names_filter = full_info_subset[(full_info_subset.Entry.str.len() == 6)]
```

```python
names_filter
```

```python
selected = names_filter[0:50]
```

```python
selected
```

```python
starting_selected = starting_table.copy()
```

```python
starting_selected['selected'] = starting_selected['Entry'].isin(set(selected['Entry'])).astype(object)
```

```python
starting_selected[starting_selected['selected'] == True]
```

```python
starting_selected
```

```python
starting_selected.to_excel('selected.xlsx')
```
